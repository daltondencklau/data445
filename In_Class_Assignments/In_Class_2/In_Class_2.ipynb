## Importing dataset from an S3 bucket

```python
import boto3
import pandas as pd; pd.set_option('display.max_column', 100)
import numpy as np


## define bucket in which you are trying to reach
s3 = boto3.resource('s3')
bucket_name = 'daltondencklau-data445-bucket'
bucket = s3.Bucket(bucket_name)


## define csv file to read in the bucket
file_key= 'framingham.csv'

## syntax to allow us to read the file
bucket_object = bucket.Object(file_key)
file_object = bucket_object.get()
file_content_stream = file_object.get('Body')

## reading the data file
heart = pd.read_csv(file_content_stream)
heart.head()
```

```python
heart.describe()
```

```python
## dropping observations with missing values
heart = heart.dropna()
```

```python
import matplotlib.pyplot as plt

## creating plot
plt.scatter(x = 'BMI', y = 'heartRate', data = heart)
plt.xlabel('BMI')
plt.ylabel('heart Rate')
plt.show();
```

```python
## from the above plot, we do not recognize any distinct relationship between BMI and heart rate
```

```python
plt.scatter(x = 'glucose', y = 'BMI', data = heart)
plt.xlabel('glucose')
plt.ylabel('BMI')
plt.show();
```

```python
## from the above plot, there appears to be two groups in the data. One of the groups shows an positive/
## exponential relationship between glucose and BMI.
```

## building a logistic regression model

```python
from sklearn.linear_model import LogisticRegression

## defining input and target variables (double square brackets for more than one input variable)
X = heart[['age', 'currentSmoker', 'totChol', 'BMI', 'heartRate']]
Y = heart['TenYearCHD']

## building logistic regression model
logit_md = LogisticRegression().fit(X, Y)

## extracting the estimated coefficients
logit_md.coef_

## output values correspond to the input variables in direct order in which they were included above in [[]]
```

```python
## prediciting the likelihood
logit_md.predict_proba([[50, 1, 250, 27, 85]])
```
